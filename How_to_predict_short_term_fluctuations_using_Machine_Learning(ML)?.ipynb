The next section is the degree input section where we input the degree.deg=0 means a constant, 1 means a straight line, 2 is a quadratic curve and so on.

```python
deg=3
```

```python
#Linear Input

# actualx =[1,2,3,4]
# #actualy=[3*x+1 for x in actualx]#Linear Input
# actualy=[4, 7, 10, 13]
# equation=poly.Polynomial.fit(actualx,actualy,deg=deg).convert()# Equation based on the input data
# print("Equation=",equation)#Print the generated equation
# print(actualx)
# print(actualy)
# predictionx=np.linspace(1,10,10)# values from 1 to 10. Predicted values input

# #Quadratic Input

# actualx =[1,2,3,4]
# # actualy=[2*x*x-3*x+1 for x in actualx]#Quadratic Input
# actualy=[0, 3, 10, 21]
# equation=poly.Polynomial.fit(actualx,actualy,deg=deg).convert()# Equation based on the input data
# print("Equation=",equation)#Print the generated equation
# print(actualx)
# print(actualy)
# predictionx=np.linspace(1,10,10)# values from 1 to 10. Predicted values input


#Sin Input from 0 to pi/2

# pi=math.pi
# actualx=[0,pi/6,pi/4,pi/3,pi/2]
# actualy=[np.sin(x) for x in actualx]
# [0.0, 0.5, 0.7071067811865475, 0.8660254037844386, 1.0]
# equation=poly.Polynomial.fit(actualx,actualy,deg=deg).convert()# Equation based on the input data
# print("Equation=",equation)#Print the generated equation
# print(actualx)
# print(actualy)
# predictionx=np.linspace(-2*pi,2*pi,60)# values from 1 to 10. Predicted values input


#Sin Input from 0 to pi

# pi=math.pi
# actualx=[0,pi/6,pi/4,pi/3,pi/2,2*pi/3,3*pi/4,5*pi/6,pi]
# actualy=[np.sin(x) for x in actualx]
# actualy=[0.0, 0.49999999999999994, 0.7071067811865475, 0.8660254037844386, 1.0, 0.8660254037844387, 0.7071067811865476, 0.49999999999999994, 1.2246467991473532e-16]
# equation=poly.Polynomial.fit(actualx,actualy,deg=deg).convert()# Equation based on the input data
# print("Equation=",equation)#Print the generated equation
# print(actualx)
# print(actualy)
# predictionx=np.linspace(-2*pi,2*pi,60)# values from 1 to 10. Predicted values input
# print(predictionx)

#Sin Input from 0 to 2pi

pi=math.pi
actualx=[0,pi/6,pi/4,pi/3,pi/2,2*pi/3,3*pi/4,5*pi/6,pi, 7*pi/6,5*pi/4, 4*pi/3,3*pi/2, 10*pi/6,7*pi/4, 11*pi/6,2*pi]
actualy=[np.sin(x) for x in actualx]
#actualy=[0.0, 0.49999999999999994, 0.7071067811865475, 0.8660254037844386, 1.0, 0.8660254037844387, 0.7071067811865476, 0.49999999999999994, 1.2246467991473532e-16]
equation=poly.Polynomial.fit(actualx,actualy,deg=deg).convert()# Equation based on the input data
print("Equation=",equation)#Print the generated equation
print(actualx)
print(actualy)
predictionx=np.linspace(-2*pi,2*pi,60)# values from 1 to 10. Predicted values input
print(predictionx)
```

**The outputs section**

```python
print("Input X",actualx,",Input Y",actualy,",Degree ",deg)
model = poly.polyfit(actualx, actualy, deg=deg)#Generate the model
predictedonactualx=poly.polyval(actualx,model)# Predict on the basis of the actual input
plt.plot(actualx, actualy,color="red",label="Actual Data")# Plot
plt.scatter(actualx, actualy,label="Actual Data", marker='o',color="red")# Scatter
plt.title("Actual Data")
plt.legend()
plt.show()



plt.plot(actualx, predictedonactualx,color="green",label="Predicted Data")
plt.scatter(actualx, predictedonactualx,label="Predicted Data", marker='o',color="green")
plt.title("Predicted Data")
plt.legend()
plt.show()
correlation=r2_score(actualy,predictedonactualx)
print("Correlation",correlation *100,"%")

plt.plot(actualx, actualy,color="red",label="Actual Data")
plt.scatter(actualx, actualy,label="Actual Data", marker='o',color="red")
plt.plot(actualx, predictedonactualx,color="green",label="Predicted Data")
plt.scatter(actualx, predictedonactualx,label="Predicted Data", marker='o',color="green")
plt.title("Actual v Predicted Data")
plt.legend()
plt.show()

predictiononpredictedx=poly.polyval(predictionx,model)# Predict on the basis of the prediction input

plt.plot(predictionx, predictiononpredictedx,color="green",label="Predicted Data")
plt.scatter(predictionx, predictiononpredictedx,label="Predicted Data", marker='o',color="green")
plt.title("Predicted Data")
plt.legend()
plt.show()
```
