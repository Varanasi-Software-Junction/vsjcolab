**The weekday names and the month names**

```python
weekdays=["Mon ","Tue","Wed","Thu","Fri","Sat",'Sun']#Weekday names
months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]#Month names
```

```python
def isLeapYear(year):
  """
  This function returns True if the given year is a leap year and false if its not a leap year.
  """
  if year%400==0:
    return True
  if year%4==0 and year%100!=0:
    return True
    return False

def daysinMonth(month,year):
  """
  This function returns the no of days in a given month and year. It assumes that the month is in the proper range 1 to 12
  """
  if month in [4,6,9,11]:
    return 30
  if month == 2:
    if isLeapYear(year):
      return 29
    else:
      return 28
  return 31

def compare(d1,m1,y1,d2,m2,y2):
  """
  A function compare that gives -1 if D1<D2, 0 f D1=D2 and 1 if D1>D2.
  """
  if y1<y2:
    return -1
  if y1>y2:
    return 1
  if m1<m2:
    return -1
  if m1>m2:
    return 1
  if d1<d2:
    return -1
  if d1>d2:
    return 1
  return 0


def dateDifference(d1,m1,y1,d2,m2,y2):
  """
  This function finds the difference in dates between 2 dates D1 and D2 with the assumption that D1>=D2
  """
  sign=compare(d1,m1,y1,d2,m2,y2)
  if sign==0:#Dates are same return 0
    return 0
  if sign==-1:#D2>D1, interchange D1 and D2 and return with -ve sign
    d1,d2=d2,d1
    m1,m2=m2,m1
    y1,y2=y2,y1

  difference1=d1-1
  difference2=d2-1#Days difference
  difference3=0
  for i in range(m1):#Months difference
    difference3+=daysinMonth(i,y1)
  difference4=0
  for i in range(m2):
    difference4+=daysinMonth(i,y2)
  difference5=0
  for i in range(y2,y1):#Years difference
    if isLeapYear(i):
      difference5+=366
    else:
      difference5+=365
  total=difference1-difference2+difference3-difference4+difference5
  return sign*total


def weekDayNo(d,m,y):
  """
  Find the weekday no with Mon=0, Tuesday=1,,,Sunday=6 for any given day
  """
  diff=dateDifference(d,m,y,5,12,2022)
  diff=diff%7
  diff=7+diff
  diff=diff%7
  return diff


def weekDayName(d,m,y):
  """
  Return the week day name for a given a date
  
  """
  day=weekDayNo(d,m,y)
  return weekdays[day]


def printCalendar(m,y):
  """
  This function draws the calendar of a month given the month no and the year.
  """
  noofdays=daysinMonth(m,y)
  startweekday=weekDayNo(1,m,y)
  for day in weekdays:#Print all the week days
    print(f'{day:4s}',end="")
  print()

  for i in range(1,startweekday+1):# Shift the cursor to the startweekday.
    print(f'{"":4s}',end="")
  for i in range(1,noofdays+1): #Print the day numbers
    print(f'{str(i):4s}',end="")
    if (i+startweekday)%7==0: #If seven printed then break the line.
      print()
```
