```python
from google.colab import files
import pandas as pd
import io
import numpy
#Read from hard disk

filesuploaded = files.upload()
df = pd.read_csv(io.BytesIO(filesuploaded['marks.csv']))

#Upload from github
"""
url = 'https://raw.githubusercontent.com/Varanasi-Software-Junction/vsjcolab/main/marks.csv'
df = pd.read_csv(url)
"""
print(df)
from sklearn import linear_model


from numpy.testing import rundocs


marks = df[['Phy', 'Chemistry']]
y = df['Percent']
print(y)
print(marks)
regvalue = linear_model.LinearRegression()
regvalue.fit(marks.values, y)



predictedtotal = regvalue.predict([[10,20]])

print("Coeff=",regvalue.coef_)

print(predictedtotal)
```
